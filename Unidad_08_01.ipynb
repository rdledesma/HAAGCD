# Unidad 6: Introducción al aprendizaje automático


Aunque Arthur Samuel no fue el primero en publicar un artículo que empleara el término "aprendizaje automático", se le atribuye la creación y definición de este concepto como el campo especializado que conocemos hoy. En su trabajo `Algunos estudios sobre aprendizaje automático usando el juego de damas', presentó el aprendizaje automático como una rama de la informática que permite a las computadoras mejorar su desempeño sin necesidad de ser programadas de forma explícita.

Si bien la definición original de Samuel no lo menciona directamente, un aspecto fundamental del aprendizaje automático es el autoaprendizaje. Este concepto implica el uso de modelos estadísticos para identificar patrones y optimizar el rendimiento con base en datos e información empírica, sin requerir instrucciones de programación explícitas.

La clasificación del aprendizaje automático (Machine Learning) se puede hacer desde varias perspectivas, pero generalmente se organiza en tres grandes tipos principales según la forma en que el modelo aprende de los datos:
1. Aprendizaje supervisado
2. Aprendizaje No supervisado

## Aprendizaje supervisado
    El aprendizaje supervisado es una de las formas más comunes y estudiadas del aprendizaje automático. En este enfoque, el modelo aprende a partir de un conjunto de datos etiquetado, lo que significa que cada entrada en los datos de entrenamiento viene acompañada de una respuesta correcta o etiqueta (también llamada variable objetivo o target).

Durante el proceso de entrenamiento, el modelo intenta establecer una relación funcional entre las variables de entrada (también llamadas features o características) y la salida deseada (etiqueta). El objetivo es que, una vez entrenado, el modelo pueda predecir correctamente la etiqueta de nuevas instancias no vistas.

### Tareas comunes en el aprendizaje supervisado

#### Clasificación: El objetivo es predecir una etiqueta discreta o categórica.

            Diagnóstico médico (enfermo / sano).

            Clasificación de correos (spam / no spam).

            Reconocimiento de dígitos escritos a mano (0–9).

#### Regresión: El objetivo es predecir un valor continuo.

            Predicción del precio de una vivienda.

            Estimación de la temperatura futura.

            Pronóstico de ventas.

### Etapas del aprendizaje supervisado

- Recolección y preparación de datos: Se necesita un conjunto de datos representativo con entradas y salidas conocidas.

- División del conjunto de datos: Generalmente se divide en datos de entrenamiento, validación y prueba.

- Selección del modelo: Se elige un algoritmo de aprendizaje adecuado (por ejemplo, regresión logística, redes neuronales, etc.).

- Entrenamiento: El modelo ajusta sus parámetros internos minimizando el error entre las predicciones y las etiquetas reales.

- Evaluación y ajuste: Se mide el rendimiento del modelo y se ajustan los hiperparámetros si es necesario.

- Predicción: Una vez validado, el modelo se usa para hacer predicciones sobre nuevos datos.


### Algoritmos comunes en regresión

    Regresión lineal

    Máquinas de vectores de soporte (SVM)

    Árboles de decisión y bosques aleatorios (Random Forests)

    Redes neuronales artificiales

#### Disclamer (sobre el aprendizaje supervisado)

    Generalmente ofrece alta precisión cuando se dispone de muchos datos bien etiquetados.

    Es adecuado para tareas donde la relación entre variables es conocida o inferible.


## Aprendizaje No Supervisado
    El modelo intenta encontrar estructuras  o patrones en datos sin etiquetar.
